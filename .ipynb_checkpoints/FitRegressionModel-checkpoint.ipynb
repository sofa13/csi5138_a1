```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint

from sklearn import preprocessing

# Run Experiment
bigN = [2, 5, 10, 20, 50, 100, 200]
deg = list(range(1,21))
sigmas = [0.01, 0.1, 1]
numCombinations = len(bigN)*len(deg)*len(sigmas)
```

```python
def gaussianRan(sigma, mu=0):
    # standard normal distribution of mean 0 variance 1
    # For random samples from N(\mu, \sigma^2), use
    return sigma*np.random.randn() + mu
```

```python
# test gaussian distribution
n = 10000
x = np.random.randn(n)
#x = np.random.uniform(low=0, high=1, size=(n,1))

bins = int(n/5)
plt.hist(x, bins)
plt.show()
```

```python
def functionx(x, sigma):
    return np.cos(2*np.pi*x) + gaussianRan(sigma)
```

```python
def getData(n, sigma):
    x = np.random.uniform(low=0, high=1, size=(n,1))
    y = functionx(x, sigma)
    
    return x, y
```

```python
# graph test
n = 100
sigma = 1
x, y = getData(n, sigma)

plt.scatter(x, y)
plt.show()
```

```python
def forward(x, w, b):
    return np.dot(x,w) + b
```

```python
def getMSE(y_pred, y):
    mse = np.mean((y_pred - y)**2)
    return mse
```

```python
def getCost(x, y, w, b):
    y_pred = forward(x,w,b)
    cost = getMSE(y_pred, y)
    return cost
```

```python
def polynomialx(x, d):
    poly = preprocessing.PolynomialFeatures(d, include_bias=False)
    x = poly.fit_transform(x)  
    return x
```

```python
x = np.array([[1], [2], [3]])
d = 4
print(polynomialx(x,d))
```

```python
def gradient(x, y, w, b):
    # as implemented in vectorized implementation by Andrew Ng
    m = x.size
    z = forward(x,w,b)
    a = z
    dz = a-y
    dw = (2/m)*np.dot(x.T,dz)
    db = (2/m)*np.sum(dz)
    return dw, db
```

```python
def gradApprox(x, y, w, b, e=1e-7):
    dwa = np.zeros(w.size).reshape(-1, 1)
    dba = 0
    
    for i in range(w.size):
        thetaPlus, thetaMinus = np.copy(w), np.copy(w)
        
        thetaPlus[i] += e
        thetaMinus[i] -= e
        
        jp = getCost(x, y, thetaPlus, b)
        jm = getCost(x, y, thetaMinus, b)
        
        dwa[i] = (jp - jm)/(2*e)
        
    dba = (getCost(x, y, w, (b+e)) - getCost(x, y, w, (b-e)))/(2*e)
    
    return dwa, dba
```

```python
def fitData(x, y, d, sigma, learning_rate=5e-2, n_test=1000, itera=1000, check_grad=False):
    
    # initialize weights
    b = 0
    w = np.zeros((d,1))
    
    # create x dimension d
    x = polynomialx(x, d)
    
    # keep track of cost for testing
    cost_hist = []
    i_hist = []
    
    for i in range(itera):
        # Compute cost for testing
        cost = getCost(x,y,w,b)
        cost_hist.append(cost)
        i_hist.append(i)
        
        # Compute derivatives
        dw, db = gradient(x,y,w,b)
        
        # grad checking
        if check_grad and i > 20:
            dwa, dba = gradApprox(x,y,w,b)
            print(dwa)
            print(dba)
            print(dw)
            print(db)
            db = np.array([db for i, x in enumerate(dw)]).reshape(dw.size,1)
            dba = np.array([dba for i, x in enumerate(dwa)]).reshape(dwa.size,1)
            dtheta = np.r_[db, dw]
            dthetaa = np.r_[dba, dwa]
            e = np.linalg.norm(dtheta-dthetaa)/(np.linalg.norm(dtheta) + np.linalg.norm(dthetaa))
            print("Grad approx error:" , e)
            break
        
        # update gradient descent
        w -= learning_rate*dw
        b -= learning_rate*db
    
    #print(cost_hist)
    
    # Calculate Ein
    Ein = getCost(x,y,w,b)
    
    # Calculate Eout, with test dataset
    x_test, y_test = getData(n_test, sigma)
    
    # create x_test dimension d
    x_test = polynomialx(x_test, d)
    
    # Calculate Eout
    Eout = getCost(x_test, y_test, w, b)
    
    return Ein, Eout, w, b, [i_hist, cost_hist]
```

```python
#x = np.array([[1],[2],[3]])
d = 19
sigma = 1
n = 2
x, y = getData(n,sigma)
Ein, Eout, w, b, plot_data = fitData(x, y, d, sigma, check_grad=False)
print(Ein)
print(Eout)

xx, yy = plot_data
plt.scatter(xx, yy)
plt.show()
```

```python
def experiment(n, d, sigma, trials=20, n_test=2000):
    # initialize
    Ein_hist, Eout_hist = np.array([]), np.array([])
    w_hist, b_hist = np.empty([d, 0]), np.array([])
    
    for trail in range(trials):
        x, y = getData(n, sigma)
        Ein, Eout, w, b = fitData(x, y, d, sigma)
        Ein_hist = np.append(Ein_hist, [Ein])
        Eout_hist = np.append(Eout_hist, [Eout])
        w_hist = np.append(w_hist, w, axis=1)
        b_hist = np.append(b_hist, [b])
    
    #print(Ein_hist)
    #print(Eout_hist)
    #print(w_hist)
    #print(b_hist)
    
    # take the avg
    Ein_avg = np.mean(Ein_hist)
    Eout_avg = np.mean(Eout_hist)
    w_avg = (1/trials)*np.sum(w_hist, axis=1).reshape(d,1)
    b_avg = np.mean(b_hist)
    
    #print(w_avg)
    #print(b_avg)
    
    # Test dataset on averaged polynomials
    x_test, y_test = getData(n_test, sigma)
    
    # create x_test dimension d
    x_test = polynomialx(x_test, d)
    
    Ebias = getCost(x_test, y_test w_avg, b_avg)
    
    return Ein_avg, Eout_avg, Ebias
```

```python
# Run Experiment
bigN = [2, 5, 10, 20, 50, 100, 200]
deg = list(range(1,21))
sigmas = [0.01, 0.1, 1]

# testing
#bigN = [50]
#deg = [4]
#sigmas = [0.1]

experiments = []

cntCombinations = 0
print("Number combinations: ", numCombinations)

for n in bigN:
    for d in deg:
        for sigma in sigmas:
            cntCombinations+=1
            Ein_avg, Eout_avg, Ebias = experiment(n, d, sigma)
            
            print("Experiment ", cntCombinations)
            print("Inputs N, d, sigma: %s, %s, %s" % (n, d, sigma))
            print("%s, %s, %s" % (Ein_avg, Eout_avg, Ebias))
            experiments.append([n, d, sigma, Ein_avg, Eout_avg, Ebias])

# save experiment
with open("experiments_5", 'w', encoding="utf-8") as fout:
    pprint(experiments, fout)
```
