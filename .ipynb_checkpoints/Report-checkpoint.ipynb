# Read and process experiments

```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
bigN = [2, 5, 10, 20, 50, 100, 200]
deg = list(range(0,21))
deg = list(range(1,21))
sigmas = [0.01, 0.1, 1]
numCombinations = len(bigN)*len(deg)*len(sigmas)
```

```python
def plot_report_graph(title, data):
    fig = plt.figure(figsize=(12, 3))
    fig.suptitle(title, fontsize=20)
    for i, entry in enumerate(data):
        data_dict = entry[0]
        x = entry[1]
        x_label = entry[2]
        bottom, top = entry[3]
        
        y_data = []
        y_data.append([[data_dict[n][0] for n in x], "Ein"])
        y_data.append([[data_dict[n][1] for n in x], "Eout"])
        y_data.append([[data_dict[n][2] for n in x], "Ebias"])
        
        plt.subplot(1,3,i+1)
        for entry in y_data:
            y, labl = entry
            plt.plot(x, y, label=labl)
        plt.xlabel(x_label)
        plt.legend(loc="upper left")
        plt.ylim((bottom, top))

    plt.show()
```

```python
def process_experiments(experiments):
    n_dict, d_dict, sigma_dict = dict(), dict(), dict()
    n_dict_cst, d_dict_cst, sigma_dict_cst = dict(), dict(), dict()
    n_dict_cst2, d_dict_cst2, sigma_dict_cst2 = dict(), dict(), dict()
    for exp in experiments:
        n, d, s = exp[0], exp[1], exp[2]

        # Averages over all variables
        n_dict[n] = n_dict.get(n, np.array([0, 0, 0])) + np.array(exp[3:6])
        d_dict[d] = d_dict.get(d, np.array([0, 0, 0])) + np.array(exp[3:6])
        sigma_dict[s] = sigma_dict.get(s, np.array([0, 0, 0])) + np.array(exp[3:6])

        # Holding other variables constant
        if d == 4 and s == 0.01:
            n_dict_cst[n] = n_dict_cst.get(n, np.array([0, 0, 0])) + np.array(exp[3:6])
        if n == 100 and s == 0.01:
            d_dict_cst[d] = d_dict_cst.get(d, np.array([0, 0, 0])) + np.array(exp[3:6])
        if n == 200 and d == 4:
            sigma_dict_cst[s] = sigma_dict_cst.get(s, np.array([0, 0, 0])) + np.array(exp[3:6])
            
        if d == 4 and s == 1:
            n_dict_cst2[n] = n_dict_cst2.get(n, np.array([0, 0, 0])) + np.array(exp[3:6])
        if n == 100 and s == 1:
            d_dict_cst2[d] = d_dict_cst2.get(d, np.array([0, 0, 0])) + np.array(exp[3:6])
        if n == 2 and d == 20:
            sigma_dict_cst2[s] = sigma_dict_cst2.get(s, np.array([0, 0, 0])) + np.array(exp[3:6])

    # Average
    n_dict = {k: x/(numCombinations/len(n_dict)) for k, x in n_dict.items()}
    d_dict = {k: x/(numCombinations/len(d_dict)) for k, x in d_dict.items()}
    sigma_dict = {k: x/(numCombinations/len(sigma_dict)) for k, x in sigma_dict.items()}
    return [n_dict, d_dict, sigma_dict, n_dict_cst, d_dict_cst, sigma_dict_cst, n_dict_cst2, d_dict_cst2, sigma_dict_cst2]
```

```python
# Read experiment file without reg
with open(".\experiments\experiments_lambd_-1e-05", "r", encoding='utf-8') as f:
    experiments = eval(f.read())
res = process_experiments(experiments) 

with open(".\experiments\experiments_lambd_0", "r", encoding='utf-8') as f:
    experiments = eval(f.read())
res_reg = process_experiments(experiments) 

with open(".\experiments\experiments_lambd_1e-05", "r", encoding='utf-8') as f:
    experiments = eval(f.read())
res_reg_1 = process_experiments(experiments) 
```

# Results
The following results show the fitting and generalization of our model, in relation to sample size (N), model complexity (d), and noise level (sigma). We also include comparisons of our model with negative and positive weight decay, as well as zero weight decay. 

First we discuss our findings when using weight decay regularization on our model. Then we dive into our plots, where we first show results when averaging combination of parameters, and next we show results when keeping two of the three parameters constant. In the section where we keep parameters constant, we look at two scenarios where the parameters that are kept constant are picked to produce low errors in one scenario and high errors in the other scenario.

Note in our results we refer to sample size (N) which take on the values {2,5,10,20,50,100,200}, model complexity polynomial degree (d) {0,1,...20}, and noise level (sigma) {0,0.01,1}. We also refer to errors Ein (training error), Eout (testing error), and Ebias.

## Weight decay regularization
Overall we find that negative weight decay decreases errors Eout, Ein and Ebias of our model. All errors decreasing shows that negative weight decay improves the fitting of our model. All errors decreasing also results in closing the gap between training and testing error, which improves the generalization of our model. The use of weight decay regularization is meant to find a balance between model complexity and generalizing well to unseen data. This is done by introducing a constant which "weighs" on the learned parameters, forcing the parameters to become smaller if the weight decay is positive, and bigger if the weight decay is negative. Therefore complex models can be made "simpler" by introducing positive weight decay, since the learned parameters are forced to be smaller. 

The tradeoff between model complexity and generalizing well to unseen data usually results in a tradeoff between training error and test error. Less model complexity usually results in better generalization, meaning training error increases while test error decreases. Conversely, greater model complexity usually results in less generlization, meaning training error decreases while test error increases. However, what we found worked best for our model was to use weight decay to increase our model complexity (by using a negative weight decay), which improved both our training error and test error. The reason for this result may be because the model complexity never becomes complex enough, due to the x's being between 0 and 1. When the model complexity degree d increases and x is taken to the power of d, these values become smaller and smaller, instead of bigger and bigger if x's were greater than 1. Therefore in order to have a more complex model, a learned model needs to learn larger parameters to offset x's very small values. Negative weight decay ecourages learned parameter "weights" to become larger, therefore increasing model complexity and fitting the data better, both for training data and test data.

## Averaging combinations of parameters
The plots below show three experiments of our model, each experiment with a different "weight decay". For each weight decay experiment, we look at the averages of Ein, Eout, and Ebias, in relation to parameters sample size (N), degree (d) and noise level (sigma). For example, in the "Weight Decay=0" experiment, looking at the plot with Sample size (N) on the x-axis, point 50 plots the average of Ein, Eout and Ebias for each parameter combination of d {0,1,...,20}, and sigma {0.01, 0.1, 1}. This gives us a rough overview of the experiments.

As an overview, we find that negative weight decay improves fitting and generalization. We also find that Ein is always lower than Eout, and that Ebias is between Ein and Eout.

Comparing Sample size (N) plots, we find Ein is best and Ebias and Eout are worst the smaller N is. This makes sense since fitting is very easy when the training set has few datapoints. Conversely, generalization will be very poor with only a few training set datapoints, since one cannot create a model that will generalize well to unseen data. As N increases, we find Ein fitting decreases and then platues. We also find the generalization gap closes as sample size (N) increases, after which all errors platue and become equal to each other the bigger N becomes. 

Comparing Model complexity (d) plots, we find all errors are worse the smaller model complexity (d) is. This makes sense since with small model complexity, such as degree d=0, the model can only fit the data with a straight line through the y-intercept, and cannot fit more complex training data. We find all the errors begin decreasing up until about degree 4, afterwhich Ein and Ebias remain rather constant, whereas Eout increases and becomes more scattered. Interestingly we find the generalization gap remains more or less constant and the gap does not consistently increase or decrease as complexity increases.

Comparing Noise level (sigma) plots, we find not too surprisingly that the smallest sigma gives the smallest errors overall. This makes sense since larger noise levels will make it harder to fit training data. We find increasing noise level sigma increases all errors, therefore resulting in underfitting, and that the generalization gap between Ein and Eout increases as sigma increases.

When comparing weight decay plots we notice all errors reduce, and therefore fitting improves, due to negative weight decay. We also notice that the generalization gaps become smaller as weight decay decreases. For example, going from the plots in experiment "Weight Decay=+1e-05" to "Weight Decay=-1e-05", in the Sample size (N) plots the generalization gap closes sooner from N=100 to N=50, in the complexity (d) plots the generalization gap reduces from magnitude 3 to magnitude 1, and in the Noise level (sigma) plots at sigma equal to 1 the generalization gap reduces from magnitude 6 to magnitude 2.

```python
plot_report_graph("Weight Decay=+1e-05", [[res_reg_1[0], bigN, "Sample size (N)", [0,4]], [res_reg_1[1], deg, "Model complexity (d)", [0,5]], [res_reg_1[2], sigmas, "Noise level (sigma)", [0,8]]])
plot_report_graph("Weight Decay=0", [[res_reg[0], bigN, "Sample size (N)", [0,4]], [res_reg[1], deg, "Model complexity (d)", [0,5]], [res_reg[2], sigmas, "Noise level (sigma)", [0,8]]])
plot_report_graph("Weight Decay=-1e-05", [[res[0], bigN, "Sample size (N)", [0,4]], [res[1], deg, "Model complexity (d)", [0,5]], [res[2], sigmas, "Noise level (sigma)", [0,8]]])
```

## Keeping parameters constant
We also want to look at the parameter plots while keeping the other parameters constant. For example, in the plots below, for "Weight Decay=0" experiment, looking at the plot with Sample size (N) on the x-axis, point 50 plots the Ein, Eout and Ebias for model complexity d=4, and noise level s=0.01. Holding the rest of the parameters constant allows us to see how Ein, Eout, and Ebias changes while only changing one parameter. 

#### Changes in sample size and model complexity with low noise level, and changes in noise level with stable sample size and degree

In this result we are particularly interested in seeing how sample size and model complexity effect model fitting and generalization with low noise (s=0.01), as well as how noise levels effect the changes in errors when sample size and model complexity are stable. Note we determined stable sample size and model complexity based on the plots above which yielded smaller errors overall when sample size n=200 and when model complexity d=4. It makes sense that these parameters yield more stable results, since more data usually results in better fitting, and our ground truth function which takes on the curve of a cosine function between 0 and 1 would be best fit by a polynomial of degree 4.

Generally we find the plots below are similar in shape to the plots above which averaged the errors for each parameter combination. However we see that with very low noise, in the sample size (N) plots the errors are smaller (fitting improves) and generalization gap closes much sooner than in the plots above. In the Model complexity (d) plots, the generalization gaps are almost non existent, and interestingly we see a more noticable improvement in all errors as complexity increases. Similarly, looking at the Noise level (sigma) plot with stable sample size and degree, we find the generalization gap is almost non existent.

```python
plot_report_graph("Weight Decay=+1e-05", [[res_reg_1[3], bigN, "Sample size (N), d=4, s=0.01", [0,5]], [res_reg_1[4], deg, "Model complexity (d), n=100, s=0.01", [0,5]], [res_reg_1[5], sigmas, "Noise level (sigma), n=200, d=4", [0,5]]])
plot_report_graph("Weight Decay=0", [[res_reg[3], bigN, "Sample size (N), d=4, s=0.01", [0,5]], [res_reg[4], deg, "Model complexity (d), n=100, s=0.01", [0,5]], [res_reg[5], sigmas, "Noise level (sigma), n=200, d=4", [0,5]]])
plot_report_graph("Weight Decay=-1e-05", [[res[3], bigN, "Sample size (N), d=4, s=0.01", [0,5]], [res[4], deg, "Model complexity (d), n=100, s=0.01", [0,5]], [res[5], sigmas, "Noise level (sigma), n=200, d=4", [0,5]]])
```
